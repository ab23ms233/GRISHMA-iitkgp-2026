Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Variance of dot product values

```python
num = 1000	# number of dot products
dims = [10, 100, 1000]	# dimesnions of vectors
dots = np.zeros(shape=num)	# array to store the dot product values
colors = ['blue', 'orange', 'green']	# colors of various histogram plots

fig = plt.figure(figsize=(14, 6))

# Plot dot product histogram for all dimensions
for i in range(len(dims)):
	# Perform dot product num number of times
	for j in range(num):
		# Generating random vectors
		np.random.seed(j)
		vec1 = np.random.random(size=(1, dims[i]))
		vec2 = np.random.random(size=(1, dims[i])).T
		# Standardising the vectors (mena=0 and variance=1)
		vec1_std = (vec1 - vec1.mean())/vec1.std()
		vec2_std = (vec2 - vec2.mean())/vec2.std()

		dot_product = np.dot(vec1_std, vec2_std)	# Dot product
		dots[j] = dot_product
	
	var = dots.var()

	# Plotting histogram
	plt.subplot(1, 3, i+1)
	plt.hist(dots, bins=30, color=colors[i], edgecolor='black', linewidth=1, label=f"Variance = {var: 0.2f}")
	plt.title(f"Dimensions = {dims[i]}")
	plt.legend()
	plt.grid()

# Title and labels
fig.supxlabel("Dot product value")
fig.supylabel("Frequency")
fig.suptitle("Histogram of dot product between vectors of various dimensions")
plt.tight_layout()
plt.show()
```

Nature of softmax function

```python
def softmax(l):
    s = np.exp(l)/np.sum(np.exp(l))
    return s

l1 = [1, 2]
l2 = [10, 20]

print("Softmax of l1:", softmax(l1))
print("Variance of l1:", np.var(l1))
print()

print("Softmax of l2:", softmax(l2))
print("Variance of l2:", np.var(l2))
```
