# Setting up the environment

## Importing libraries

```python
import pandas as pd
import numpy as np
```

## Contructing dataframe of features and activations

```python
# Features
f1 = np.array([2, 1, 5, 6, 7])
f2 = np.array([3, 1, 4, 1, 1])

# Activations
z1 = np.array([7, 2, 1, 7, 3])
z2 = np.array([5, 3, 2, 5, 3])

# Dataframe
df = pd.DataFrame(data={'f1': f1, 'f2': f2, 'z1': z1, 'z2': z2})
df
```

# Batch Normalisation

## Step 1: Normalisation

```python
epsilon = 1e-4
dfb = df.copy()

# Calculating mean and standard deviation
m1 = dfb['z1'].mean()
s1 = dfb['z1'].std()
m2 = dfb['z2'].mean()
s2 = dfb['z2'].std()

# Normalisation
dfb['z1'] = (dfb['z1']-m1)/(s1+epsilon)     # To avoid zero division error
dfb['z2'] = (dfb['z2']-m2)/(s2+epsilon)
dfb
```

## Step 2: Scaling and shifting

```python
# Some value of gamma and beta
gamma1, beta1 = 0.8, 0.5
gamma2, beta2 = 0.7, 0.4

# Scaling and shifting
dfb['z1'] = gamma1*dfb['z1'] + beta1
dfb['z2'] = gamma2*dfb['z2'] + beta2
dfb
```

## **Step 3: Applying activation function**. We consider the activation function to be relu.
$$RelU = max(0, z)$$

```python
def relu(z):
    # Using np.maximum for efficient element-wise ReLU
    return np.maximum(0, z)

# Calculating activations
a1 = relu(dfb['z1'])
a2 = relu(dfb['z2'])

dfb['a1'] = a1
dfb['a2'] = a2
dfb
```

# Layer Normalisation

## Step 1: Normalisation

```python
dfl = df.copy()

for i in range(len(dfl)):
    # Ensure the columns are of float type before assignment
    dfl['z1'] = dfl['z1'].astype(float)
    dfl['z2'] = dfl['z2'].astype(float)

    # Calculating mean and variance
    data = np.array(dfl.loc[i])[-2:]
    mean = data.mean()
    std = data.std()

    # Normalisation
    data = (data-mean)/(std+epsilon)    # To avoid zero division error
    dfl.iloc[i, -2:] = data

dfl
```

## Step 2: Scaling and shifting

```python
# Some value of gamma and beta
gamma1, beta1 = 0.8, 0.5
gamma2, beta2 = 0.7, 0.4

# Scaling and shifting
dfl['z1'] = gamma1*dfl['z1'] + beta1
dfl['z2'] = gamma2*dfl['z2'] + beta2
dfl
```

## **Step 3: Applying activation function**. We consider the activation function to be relu.
$$RelU = max(0, z)$$

```python
def relu(z):
    # Using np.maximum for efficient element-wise ReLU
    return np.maximum(0, z)

# Calculating activations
a1 = relu(dfl['z1'])
a2 = relu(dfl['z2'])

dfl['a1'] = a1
dfl['a2'] = a2
dfl
```
